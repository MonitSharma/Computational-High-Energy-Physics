## Plots

Plot the $\chi$ data with respect to the temperature limit.

All Chi data

```python
import numpy as np
from matplotlib import pyplot as plt

data = np.loadtxt('All_Chis.dat')
for i in range(1,22):
  X = data[:,0]
  Y = data[:,i]
  plt.xlabel("Temperature")
  plt.ylabel("$\chi$")
  plt.plot(X, Y, color='m', marker='.', linestyle='dashed',
      linewidth=2, markersize=6)
  plt.show() 
```
